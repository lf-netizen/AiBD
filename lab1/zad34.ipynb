```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Zadanie 3

```python
def f(x): return x**2 + 5
```

```python
for start, end in [(-1, 1), (-6, 6), ( 0, 5)]:
    x = np.linspace(start, end, 1000)
    y = f(x)
    plt.figure(figsize=(4, 4))
    plt.plot(x, y)
    plt.title(f'Wykres funkcji f(x) = x^2+5 w przedziale ({start}, {end})')
    plt.xlabel('x')
    plt.ylabel('y = f(x)')
    plt.legend('f(x)')
    plt.show()
```
